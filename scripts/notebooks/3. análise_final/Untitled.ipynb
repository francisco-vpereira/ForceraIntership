```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x1 = np.linspace(0,4,100)
x2 = np.linspace(4,8,100) 
```

```python
def f1(x):
    d = np.exp(x-4) - 3*x + 11
    return d/(4-x)


def f2(x):
    return np.log(2*np.exp(x) - np.exp(4))
```

```python
plt.plot(x1,f1(x1))
plt.plot(x2,f2(x2))
```
